# 30 short puzzles on probability

Many probability distributions are related to each other via simple transformations or (re)parametrizations. This notebook challenges you to derive 30 more or less known distributions from a set of just 4: [Normal](https://www.pymc.io/projects/docs/en/stable/api/distributions/generated/pymc.Normal.html), [Gamma](https://www.pymc.io/projects/docs/en/stable/api/distributions/generated/pymc.Gamma.html), [Beta](https://www.pymc.io/projects/docs/en/stable/api/distributions/generated/pymc.Beta.html), and [Student-T](https://www.pymc.io/projects/docs/en/stable/api/distributions/generated/pymc.Studentt.html).

Under the hood, it relies on PyMC [Automatic Probability](https://www.youtube.com/watch?v=0B3xbrGHPx0) to check if the probability of a random variable graph matches the target distribution. You don't need to be familiar with PyMC to solve these exercises, but you may need to check the documentation of the 4 basic distributions to see how they are parametrized. Just click on the links above to get there.

Variables from these 4 distributions can be further transformed with any operations in the [math](https://www.pymc.io/projects/docs/en/stable/api/math.html) module to solve each problem. Your task is to figure out exactly which operations and parametrizations are needed to obtain the target distribution! Google and Wikipedia are your friends!

*Note: Not all operations are listed in the docs. In general, for every `np.foo` there should be an equivalent `math.foo` that you can use. For example for a `np.log` you can use `math.log`.*

Now we need to define a Log-Normal. We can't simply use the handy `pymc.LogNormal`. That would be too easy!

```python
def dist(mu, sigma):
    return pymc.LogNormal.dist(mu, sigma)

try:
    test(dist, exercise="Log-Normal")
except Exception as exc:
    print(exc)
```

A Log-Normal is an exponentiated Normal variable. We can use the `exp` in the `math` module to define that distribution.

```python
def dist(mu, sigma):
    return math.exp(Normal(mu, sigma))

test(dist, exercise="Log-Normal")
```
